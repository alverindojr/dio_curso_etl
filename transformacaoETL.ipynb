```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['**','###!','####','****','*****','NULL']
df = pd.read_csv("ocorrencia.csv",parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_cidade":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo":pa.Column(pa.String, nullable=True),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes":pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```

```python
#ocorrencias com mais de 10 recomendacoes
filtro = df.total_recomendacoes > 10
df.loc[filtro, ['ocorrencia_cidade','total_recomendacoes']]
```

```python
#ocorrencia com a classificacao de incidente grave
filtro = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
df.loc[filtro]
```

```python
#ocorrencias com a classificao == 'INCIDENTE GRAVE' e o estado == 'SP'
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
#ocorrencias onde a classificacao == 'INCIDENTE' ou o estado == 'SP'
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 | filtro2]
```

```python
#ocorrencias onde a classificacao == 'INCIDENTE GRAVE' ou a classificacao == 'INCIDENTE' e estado == 'SP'
filtro1 = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE') 
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
#ocorrencias onde a classificacao == 'INCIDENTE GRAVE' ou a classificacao == 'INCIDENTE' e estado == 'SP'
filtro1 = df.ocorrencia_classificacao.isin (['INCIDENTE GRAVE','INCIDENTE']) 
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
#trazer todas as ocorrencias onde a cidade comeca com a letra == 'C'
filtro = df.ocorrencia_cidade.str[0] == 'C'
df.loc[filtro]
```

```python
#trazer todas as ocorrencias onde a cidade finaliza com a letra == 'A'
filtro = df.ocorrencia_cidade.str[-1] == 'A'
df.loc[filtro]
```

```python
#ocorrencias onde a cidade finaliza com os caracteres == 'MA'
filtro = df.ocorrencia_cidade.str[-2:] == 'MA'
df.loc[filtro]
```

```python
#ocorrencias onde a cidade contem os caracteres == 'MU'
filtro = df.ocorrencia_cidade.str.contains('MU')
df.loc[filtro]
```

```python
#ocorrencias onde a cidade contem os caracteres == 'MU' ou 'CA'
filtro = df.ocorrencia_cidade.str.contains('MU|CA')
df.loc[filtro]
```

```python
#ocorrencias onde o ano é == 2010
filtro = df.ocorrencia_dia.dt.year == 2010
df.loc[filtro]
```

```python
#ocorrencias onde o ano == 2010 e mes == 12 e dia == 30
filtro_ano = df.ocorrencia_dia.dt.year == 2010
filtro_mes = df.ocorrencia_dia.dt.month == 12
filtro_dia = df.ocorrencia_dia.dt.day == 30
df.loc[filtro_ano & filtro_mes & filtro_dia]
```

```python
#concatenar as colunas ocorrencia_dia + ocorrencia_hora em uma unica coluna 
#observando que uma é do tipo data e a outra do tipo string
#primeiro convertemos a coluna ocorrencia_dia que é data para string, assim as duas strings se juntam
#depois convertemos todo o conteudo string para data
#por ultimo jogamos a informacao dentro da nova coluna
df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
df.head()
```

```python
#ocorrencias entre os periodos 2010-01-03 03:00:00 as 2010-01-03 11:05:00
filtro1 = df.ocorrencia_dia_hora >= '2010-01-03 03:00:00'
filtro2 = df.ocorrencia_dia_hora <= '2010-01-03 11:05:00'
df.loc[filtro1 & filtro2]
```

```python
#criacao de um novo dataframe com o resultado do filtro do ano de 2015 e mes 03
filtro_ano = df.ocorrencia_dia.dt.year == 2015
filtro_mes = df.ocorrencia_dia.dt.month == 3
df201503 = df.loc[filtro_ano & filtro_mes]
df201503
```
